# นำเข้าข้อมูล

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
from sklearn import preprocessing
from collections import Counter
stdsc = preprocessing.StandardScaler()
```

```python
df = pd.read_csv('indian_food.csv')
```

```python
df.ingredients = df.ingredients.str.lower()
df.flavor_profile = df.flavor_profile.replace('-1', 'other')
df.state = df.state.replace('-1', 'other')
df.region = df.region.replace('-1', 'Other')
df.region = df.region.fillna('Other')
```

### - จำนวนครั้งของรายชื่อส่วนผสมที่ปรากฏ

### - จัดรูปแบบข้อมูล

```python
df = df.set_index('name')
```

```python
food_ingredient =df.ingredients.str.split(', ', expand=True).reset_index().melt(id_vars=['name'])
food_ingredient = food_ingredient.drop('variable', axis=1)
food_ingredient.columns = ['name', 'ingredient']
food_ingredient = food_ingredient.dropna()
```

```python
food_ingredient
```

```python
pd.get_dummies(food_ingredient)
```

```python
food_ingredient_hot_encode = food_ingredient.pivot_table(index=['name'], columns=['ingredient'], aggfunc=[len], fill_value=0).droplevel(0, axis=1)
```

```python
food = df.loc[:,['diet', 'flavor_profile']].join(food_ingredient_hot_encode)
```

```python
food
```

# เรียนรู้ข้อมูล และทำนายผล

```python
from sklearn.model_selection import train_test_split
```

```python
X = pd.get_dummies(food)
y = df.course
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=45)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 
 'max_depth' : [2,3,4,5,6,7,8,9,10],
 'criterion' :['gini', 'entropy']
}
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTree = DecisionTreeClassifier(random_state=0)
CV_DTree = GridSearchCV(estimator=DTree, param_grid=param_grid, cv= 2)
CV_DTree.fit(X_train, y_train)
```

```python
CV_DTree.best_params_
```

```python
DTree = DecisionTreeClassifier(random_state=0, max_depth=3, criterion='gini')
DTree.fit(X_train, y_train)
DTree.score(X_test,y_test)
```

```python
from graphviz import Source
from sklearn.tree import export_graphviz
```

```python
Source(export_graphviz(DTree, out_file=None, 
                       feature_names=X.columns, class_names=df.course.unique(),
                       filled=True, precision=2))
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB().fit(X_train,y_train)
gnb.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB().fit(X_train,y_train)
mnb.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import BernoulliNB

bnb = BernoulliNB().fit(X_train,y_train)
bnb.score(X_test,y_test)
```
